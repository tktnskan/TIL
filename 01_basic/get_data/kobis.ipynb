```python
import requests
import csv
data1 = []
f_r = open('boxoffice.csv', 'r+', encoding='utf-8-sig', newline='')
reader = csv.reader(f_r)
# http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=087ca6a1490cb65922d848fe9ea7121b&movieCd=20182935
for line in reader:
    key = ''
    url_detail = f'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={key}&movieCd={line[0]}'
    data1.append(requests.get(url_detail).json()['movieInfoResult'])
                 
data1
```

```python
movie_detail = []
for i in range(43):
    results = []
    if ',' in data1[i]['movieInfo']['movieNm']:
        data1[i]['movieInfo']['movieNm'].replace(',','.')
    results.append(data1[i]['movieInfo']['movieCd']),
    results.append(data1[i]['movieInfo']['movieNm']),
    results.append(data1[i]['movieInfo']['movieNmEn']),
    results.append(data1[i]['movieInfo']['movieNmOg']),
    results.append(data1[i]['movieInfo']['openDt']),
    results.append(data1[i]['movieInfo']['showTm']),
    results.append(data1[i]['movieInfo']['genres'][0]['genreNm']),
    results.append(data1[i]['movieInfo']['directors'][0]['peopleNm']),
    if data1[i]['movieInfo']['actors'] != []:
        num1 = 0 
        num2 = 0
        if len(data1[i]['movieInfo']['actors']) > 3:
            for j in range(3):
                results.append(data1[i]['movieInfo']['actors'][num1]['peopleNm'])
                num1 += 1
        else:
            for k in data1[i]['movieInfo']['actors']:
                results.append(data1[i]['movieInfo']['actors'][num2]['peopleNm'])
                num2 += 1
    else:
        results.append('배우 없음')
        
    movie_detail.append(results)

movie_detail
```

```python
import csv

f = open('movie.csv', 'a+', encoding='utf-8-sig', newline='')

writer = csv.writer(f)

for i in movie_detail:
    writer.writerow(i)

f.close()
```

```python
import requests
import time

movies = []
count = 0
for i in range(len(movie_detail)):
    movies.append(movie_detail[i][1])
    
naver_uri = 'https://openapi.naver.com/v1/search/movie.json?query='
client_id = ''
client_secret = ''
headers = { 
    'X-naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret 
}
data = []
for movie in movies:
    data.append(requests.get(naver_uri + movie, headers = headers).json())
    count += 1
    if count % 9 == 0:
        time.sleep(1)

data
```
